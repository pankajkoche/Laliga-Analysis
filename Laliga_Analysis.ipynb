### replace dashes with 0

### 4.Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference

```python
def  Goal_diff_count(md):
    md['GoalsFor']=md.GoalsFor.astype(float)
    md['GoalsAgainst'] = md.GoalsAgainst.astype(float)
    md['GoalsDifferences']=md['GoalsFor']-md['GoalsAgainst']
    return md
Goal_diff_count(md)
```
